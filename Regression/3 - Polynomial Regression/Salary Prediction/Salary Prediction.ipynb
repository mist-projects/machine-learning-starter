# **Step 7 : Visualize the Polynomial Regression results**

# **Problem Statement :**
Predict salaty based on position and level

# **Step  1 : Import Libraries**


```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# **Step 2 : Import Dataset**

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values 
y = dataset.iloc[:, -1].values # -1 represents last column
```

```
X.size
```

# **Step 3 : Visualize dataset**

```
dataset.head(4)
```

```
sns.jointplot(x = 'Salary', y = 'Position', data = dataset)
```

```
sns.lmplot(x = 'Level', y = 'Salary', data = dataset)
```

```
sns.lmplot(x = 'Salary', y = 'Level', data = dataset)
```

# **Step 4 : Train the Linear Regression model on the whole dataset**

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

# **Step 5 : Train the Polynomial Regression model on the whole dataset**

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
```

```
X_poly.size
```

```
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

# **Step 6 : Visualize the Linear Regression results**

```
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title = ('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# **Step 7 : Visualize the Polynomial Regression results**

```
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title = ('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# **Step 8 : Visualize the Polynomial Regression results (for higher resolution and smoother curve)**

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title = ('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# **Step 9 : Predict a new result with Linear Regression**

```
lin_reg.predict([[6.5]])
```

# **Step 10 : Predict a new result with Polynomial Regression**

```
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
