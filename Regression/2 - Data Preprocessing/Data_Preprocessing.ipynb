# **Data Preprocessing**
Handle missing values
<br>Encode categorical values

# **Step 1 - Import Libraries**

```
import numpy as np # Work with array
import matplotlib.pyplot as plt # Plotting
import pandas as pd # Work with dataset preprocessing
```

# **Step 2 - Import Dataset**

```
dataset = pd.read_csv('Data_Preprocessing.csv')
X = dataset.iloc[:, :-1].values # -1 means index of last column (excludes upper bound)
y = dataset.iloc [:, -1].values
```

```
print(X)
```

```
print(y)
```

# **Step 3 - Handle Missing Data**

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```
print(X)
```

# **Step 4 : Encode Categorical Data**

### **Encoding the Independent Variable**

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### **Encoding the Dependent Variable**

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

# **Step 5 : Split into Training and Test dataset**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1) #random_state=1 will ensure same set of train and test is produced on each run
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

# **Step 6 - Feature Scaling**

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:] = sc.fit_transform(X_train[:,3:])
X_test[:,3:] = sc.transform(X_test[:,3:]) # No fit transform here to use the same model as X_train
```

```
print(X_train)
```

```
print(X_test)
```
